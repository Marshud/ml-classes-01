```python
import pandas as pd
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price', axis='columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X, y)
```

```python
model.predict([[2800, 0, 1]])
```

```python
model.predict([[3400, 0, 0]])
```

```python
model.score(X, y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Re-create X from dfle to ensure we have the original labeled data for town and area
# as X was last defined as dfle[['town', 'area']].values
X_transformed = dfle[['town', 'area']].values

# The ColumnTransformer takes a list of (name, transformer, columns) tuples
# Here, we apply OneHotEncoder to the first column (index 0) and 'passthrough' the second column (index 1)
ct = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(), [0]), # Apply OneHotEncoder to the 'town' column (index 0)
        ('passthrough', 'passthrough', [1]) # Keep the 'area' column (index 1) as is
    ],
    remainder='drop' # Drop any other columns not specified (though not strictly necessary here as we specify all)
)

# Apply the transformation
X_new = ct.fit_transform(X_transformed)

# Display the transformed X
print("Original X:")
print(X_transformed)
print("\nTransformed X with OneHotEncoder on 'town' and 'area' passed through:")
print(X_new)
```
