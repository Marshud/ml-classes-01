```python
# Linear Regression ( A simple model to do predictions of home prices)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("./HousePricesData.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.coef_
```

```python
reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

```python
d = pd.read_csv('./Areas.csv')
d.head(5)
```

```python
p = reg.predict(d)
p
```

```python
d['prices'] = p
d
```

```python
d.to_csv('prediction.csv')
```

## Exercise on Linear Regression
Based on the Canada CSV, predict the price of houses in 2020 since the CSV ends on 2016

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('./canada_per_capita_income.csv')
df.head(10)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Income per Capita($)')
plt.scatter(df['year'], df['per capita income (US$)'], color="red", marker="+")
```

```python
reg_capita = linear_model.LinearRegression()
reg_capita.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg_capita.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Income per Capita($)')
plt.scatter(df['year'], df['per capita income (US$)'], color="blue", marker="*")
plt.plot(df['year'], reg_capita.predict(df[['year']]), color="red")
```
